```python
#Import Libraries

import pandas as pd
from datetime import datetime
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.io as pio
pio.renderers.default = 'colab'
import plotly.express as px
```

```python
#Load CSV

df = pd.read_csv('/content/Analysis.csv')
```

```python
df
```

```python
#convert date(str) into Datetime object
df['date'] = pd.to_datetime(df['date'])
```

```python
#Start and End Date in the dataFrame

df.reset_index()['date'].agg(['min', 'max'])
```

```python
#convert time(str) into Datetime object

df['time'] = pd.to_datetime(df.time).dt.strftime('%H:%M:%S')
```

```python
# No.of unique users in the DataFrame

len(df['user'].unique())
```

```python
#date, user, occurance

dates = df['date'].unique()
users = df['user'].unique().tolist()
```

```python
dlist = []
for d in dates:
    
    temp_df = df[df['date']==d]
    for usr in users:
        l = len(temp_df[temp_df['user']==usr])
        if l==0:
            continue
        tdict = {'date': d}
        tdict['user']=usr
        tdict['occurance']=l
        dlist.append(tdict)
```

```python
occ_df = pd.DataFrame(dlist, columns = ['date', 'user', 'occurance'])
```

```python
occ_df
```

```python
#Scatter plot that shows per user req over date

fig = px.scatter(occ_df, x="date", y="occurance", color="user")
fig.show()
```
